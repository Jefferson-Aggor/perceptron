```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
positive_points = np.array([(-2, 3), (0, 1), (2, -1)])
negative_points = np.array([(-2, 1), (0, -1), (2, -3)])

positive_points[:, 1]
```

```python
plt.scatter(positive_points[:, 0], positive_points[:, 1], label='Positive Values')
plt.scatter(negative_points[:, 0], negative_points[:, 1], label='Negative Values')
plt.legend(loc='best')
plt.title("Initial data points")
plt.xlabel('x1')
plt.ylabel('x2')
plt.grid(True)
plt.show()
```

```python
X = np.vstack([positive_points, negative_points])
y = np.array([1]*len(positive_points) + [-1]*len(negative_points))
```

```python
def perceptron(X, y, k=10):
    theta = np.zeros(X.shape[1])
    theta_0 = 0

    for t in range(k):
        for i, v in enumerate(X):
            if y[i] * (np.dot(theta , v) + theta_0) <= 0:
                theta += y[i]*v
                theta_0 += y[i]
                break
        plot_boundaries(theta, theta_0, t + 1)
                
    return theta, theta_0 
```

```python

theta, theta_0 = perceptron(X, y, 10)
```

```python
theta_0
```

```python
def plot_boundaries(theta, theta_0, step):
    x_vals = np.linspace(-3, 3, 9)

    if theta[1] != 0:
        y_vals = -((theta[0] * x_vals) + theta_0)/theta[1]
        plt.plot(x_vals, y_vals, color='red')
    else:
        x_val = -theta_0/theta[0]
        plt.axvline(x=x_val)

    plt.scatter(positive_points[:, 0], positive_points[:, 1], label='Positive Values')
    plt.scatter(negative_points[:, 0], negative_points[:, 1], label='Negative Values')
   
    plt.legend(loc='best')
    plt.title(f'Decision boundary at {step} iteration')
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.grid(True)
    plt.show()
```
